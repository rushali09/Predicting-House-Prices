```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_excel('data.xlsx')
df.columns = df.columns.str.strip()
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('Area(sqr ft)')
plt.ylabel('Prices(US$)')
plt.scatter(df.Area, df.Prices, color= 'red', marker= '+')
plt.show()
```

```python
# Creating object for linear regression
reg = linear_model.LinearRegression()
reg.fit(df[['Area']], df.Prices)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# Generating csv file to predict prices
d= pd.read_excel('area.xlsx')
```

```python
d
```

```python
# Using regression model to predict prices 
p = reg.predict(d)
```

```python
d['Prices'] = p
d
```

```python
d.to_excel('Price.xlsx', index= False)
```
